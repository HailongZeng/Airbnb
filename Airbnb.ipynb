# 导入数据

```
from sklearn.model_selection import train_test_split
```

```
data1 = pd.read_csv('data1.csv')
```

```
data1.columns
```

```
data1.head()
```

```
y = data1['dim_is_requested']
```

```
items = ['dim_is_requested', 'Unnamed: 0', 'ds_night', 'ds', 'id_listing_anon', 'id_user_anon']
```

```
x = data1.drop(items, axis = 1)
```

```
X = pd.get_dummies(x).reset_index(drop = True)
```

```
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 123)
```

```
X_train.shape
```

# Model

## 逻辑回归

```
from sklearn.linear_model import LogisticRegression
```

```
lr = LogisticRegression(random_state=0, solver='lbfgs', max_iter = 1000)
```

```
lr.fit(X_train, y_train)
```

```
y_pred = lr.predict(X_test)
```

```
(y_pred == y_test).sum() / y_pred.shape[0]
```

## 随机森林

```
from sklearn.ensemble import RandomForestClassifier
```

```
clf = RandomForestClassifier()
```

```
clf.fit(X_train, y_train)
```

```
y_pred = clf.predict(X_test)
```

```
(y_pred == y_test).sum() / y_pred.shape[0]
```

## GBDT

```
from sklearn.ensemble import GradientBoostingClassifier
```

```
gbdt = GradientBoostingClassifier()
```

```
gbdt.fit(X_train, y_train)
```

```
y_pred = gbdt.predict(X_test)
```

```
(y_pred == y_test).sum() / y_pred.shape[0]
```
